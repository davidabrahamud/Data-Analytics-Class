```python
# Calculate average Function
def calcAverage(scores,weights):
    avg = 0
    for i in range(0,len(scores)): # Go through the list of scores 
        avg = avg + scores[i] * weights[i] #This line specifically calculates the average
    return avg
```

```python
# Creating the dictionary of students
students = {\
            "Chris":[57,40,70,61],\
            "Nick":[100,86,74,70],\
            "Dan":[63,45,60,80],\
            "Josh":[82,91,80,86]}
```

```python
# List of weights
weights = [.1,.2,.3,.4]
```

```python
name = input("Students Name:") #Prompt the user to enter a name
scores = students[name] # Use the dictonary of students
avg = round(calcAverage(scores,weights),2) #This is used to round the average value
print(name + "'s final score is {:.2f}%.".format(avg)) 
```
